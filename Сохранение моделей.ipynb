```python
from sklearn.datasets import load_iris
from tensorflow.keras.utils import to_categorical

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

import pickle
```

```python
!pip install sklearn_porter #потирование на JAVA и C
```

```python
from sklearn_porter import Porter #потирование на JAVA и C
```

##Обучение классификатора

```python
iris = load_iris()
X = iris.data
y = iris.target
```

```python
tree = DecisionTreeClassifier()
tree.fit(X,y)
```

#Сохранение моделей

```python
with open('tree_pkl.pkl', 'wb') as f:
  pickle.dump(tree, f)
```

```python
#проверка  модели

tree.predict([X[1]])
```

```python
y[1]
```

```python
with open('tree_pkl.pkl', 'rb') as f:
  load_model = pickle.load(f)
```

```python
#загруженная модель
load_model.predict([X[1]])
```

##Портирование

```python
# Export:
porter = Porter(tree, language='java')
output = porter.export(embed_data=True)
print(output)
```

```python
# Export:
porter = Porter(tree, language='c')
output = porter.export(embed_data=True)
print(output)
```

Random forest

```python
Rf = RandomForestClassifier()
Rf.fit(X,y)

# Export C:
porter = Porter(Rf, language='c')
output = porter.export(embed_data=True)
print(output)

# Export JAVA:
porter = Porter(Rf, language='java')
output = porter.export(embed_data=True)
print(output)

#pkl
with open('Rf.pkl', 'wb') as f:
  pickle.dump(tree, f)
```
