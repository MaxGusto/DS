###Создание модели

```python
from sklearn.model_selection import train_test_split
```

```python
X_train_val, X_test, y_train_val, y_test = train_test_split(X, y, test_size=0.2, random_state=1) #деление на тест и трейн
X_train, X_val, y_train, y_val = train_test_split(X_train_val, y_train_val, test_size=0.3, random_state=1) #деление на трейн и валид
```
