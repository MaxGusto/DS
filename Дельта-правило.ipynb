```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 1, 100)
y = x * 100
plt.xlabel("Метры")
plt.ylabel("Сантиметры")
plt.plot(x, y);
```

```python
y
```

```python
w = 4

for _ in range(10000):
    i = np.random.randint(0, len(x))
    d = y[i] - w * x[i]
    w += 0.001 * w * d * x[i]
```

```python
w = 4
ws = []
ds = []

for _ in range(10000):
    i = np.random.randint(0, len(x))
    d = y[i] - w * x[i]
    w += 0.001 * w * d * x[i]
    ws.append(w)
    ds.append(d)
```

```python
plt.xlabel("Итерации обучения")
plt.ylabel("Значение веса")
plt.plot(ws);
```

```python
plt.xlabel("Итерации обучения")
plt.ylabel("Величина ошибки")
plt.plot(ds);
```

```python
plt.plot(x, y, label = "y")
plt.plot(x, w * x, label = "Предсказание")
plt.xlabel("Метры")
plt.ylabel("Сантиметры")
plt.legend();
```

```python
x = np.random.rand(10)
y = 0.5 * x + np.random.rand(10) / 5
plt.xlabel("x")
plt.ylabel("y")
plt.scatter(x, y,  label = "x");
```

```python
w = 4
ws = []
ds = []

for _ in range(10000):
    i = np.random.randint(0, len(x))
    d = y[i] - w * x[i]
    w += 0.01 * w * d * x[i]
    ws.append(w)
    ds.append(d)
```

```python
plt.plot(x, w * x, label = "Предсказание")
plt.xlabel("x")
plt.ylabel("y")
plt.scatter(x, y, label = "y")
plt.legend();
```

```python
plt.xlabel("Итерации обучения")
plt.ylabel("Значение веса")
plt.plot(ws);
```
