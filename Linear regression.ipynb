### Линейная регрессия

Y = a*x + b   
если a = 0, то Y = b

если
b = 0, то Y = a * x  


Y = a1 * x1 + a2 * x2 + ... + b

```python
from sklearn import preprocessing
```

Цена = а * площадь + b

```python
df_x = df[["Общая.площадь"]]
df_y = df[["Цена"]]
```

```python
min_max_scaler_x = preprocessing.MinMaxScaler()
min_max_scaler_y = preprocessing.MinMaxScaler()

df_scaled_x = min_max_scaler_x.fit_transform(df_x)
df_scaled_y = min_max_scaler_y.fit_transform(df_y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df_scaled_x,
                                                    df_scaled_y[:, 0],
                                                    test_size=0.2)
```

```python
y_pred_df = pd.DataFrame(y_pred)
y_pred_df = min_max_scaler_y.inverse_transform(y_pred_df)
y_test_df = pd.DataFrame(y_test)
y_test_df = min_max_scaler_y.inverse_transform(y_test_df)
```

```python
plt.scatter(y_test_df, y_pred_df)
plt.xlabel("Настоящие значения")
plt.ylabel("Предсказанные значения")
plt.show()
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test_df, y_pred_df))
print('Mean Squared Error:', metrics.mean_squared_error(y_test_df, y_pred_df))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test_df, y_pred_df)))
```

Оставим некоторые прризнаки на свое усмотрение

```python
df = pd.read_csv("data/dataset.csv")
df = df.drop("id", axis=1)
```

```python
df_y = df[["Цена"]]
df_x = df.drop(["Цена", 'Без.ремонта',
                'Дизайнерский', 'Евроремонт','Косметический',
                'Железобетонные.перекрытия', 'Смешанные.перекрытия',
                'Деревянные.перекрытия', 'Иные.перекрытия',
                'Панельные.стены', 'Блочные.стены', 'Деревянные.стены',
                'Кирпичные.стены', 'Монолитные.стены', 'Смешанные.стены'], axis=1)
```

```python
df_x.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(np.array(df_x),
                                                    np.array(df_y)[:, 0],
                                                    test_size=0.2,
                                                    random_state=0)
```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
str1 = f"Y = {regressor.intercept_} + "
for coef, col in zip(regressor.coef_, df_x.columns):
    str1 += str(np.round(coef, 3)) + " * " + str(col) + " + "
str1
```

```python
y_pred = regressor.predict(X_test)
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
